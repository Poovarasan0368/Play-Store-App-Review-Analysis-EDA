```python
# counting the null_values present in each columns

def Null_values():
  Null_values = pd.DataFrame(index= play_store.columns)
  Null_values['Non Null counts'] = play_store.notnull().sum()
  Null_values['Null counts'] = play_store.isnull().sum()
  return Null_values

Null_values()
```

**The above Dataframe show the Total counts of Null and Non Null values present in each calumns**


```python
# counting the unique values present in each colunms

uniques_counts = pd.DataFrame(index= play_store.columns)
uniques_counts['Total_counts'] = play_store.count()
uniques_counts['uniques_counts'] = play_store.nunique()

uniques_counts
```

**The above Dataframe show the Total counts of unique values present in each calumns**


```python
# counting the duplicated values present in each colunms

def Duplicated_values():
  
  temp = [play_store['App'].duplicated().sum() , play_store['Category'].duplicated().sum(), play_store['Rating'].duplicated().sum(),
          play_store['Reviews'].duplicated().sum(), play_store['Size'].duplicated().sum(), play_store['Installs'].duplicated().sum(),
          play_store['Type'].duplicated().sum(), play_store['Price'].duplicated().sum(), play_store['Content Rating'].duplicated().sum(), 
          play_store['Genres'].duplicated().sum(), play_store['Last Updated'].duplicated().sum(), play_store['Current Ver'].duplicated().sum(), 
          play_store['Android Ver'].duplicated().sum()]

  Duplicated_values = pd.DataFrame(temp,index = (play_store.columns), columns = ['Duplicated values'])
  Duplicated_values['Total_counts'] = play_store.count()
  return Duplicated_values

Duplicated_values()
```

**The above Dataframe show the Total counts of duplicated values present in each calumns**

# **5. Convert the Installs, Price and Size columns for EDA analysis**


```python
# Removing '+' and ',' Symbol from Insalls Columns

play_store['Installs'] = play_store['Installs'].str.replace('+','').str.replace(',','')
play_store.tail()
```

```python
# Removing '$' Symbol From price columns
play_store['Price'] = play_store['Price'].apply(lambda x : str(x).lstrip('$'))
```

```python
# Creating function for to convert Megabyte to kilobytes and Remove M from size column

def converting_size(i):

  if i[-1] == 'M':
      M = float(i[0:-1])
      M = M*1024
      return M
  else :
      return i
```

```python
# using the created coverting_size function
play_store['Size'] = play_store['Size'].apply(lambda x :converting_size(x))
```

```python
# Removing 'k' from Size column
play_store['Size'] = play_store['Size'].apply(lambda x : str(x).rstrip('k'))

# converting 'Varies with device' to '0' from Size column
play_store['Size'] = play_store['Size'].apply(lambda x : 0 if str(x) == 'Varies with device' else x )

# converting Size data type into numeric data type
play_store['Size'] = pd.to_numeric(play_store['Size'])

# converting Size into Megabytes
play_store['Size'] = play_store['Size'].apply(lambda x : x/1024)

# check play_store data frame 
play_store.reset_index().drop('index', axis=1)
```

```python
# converting the data frame columns to appropirate data type

play_store['Installs'] = pd.to_numeric(play_store['Installs'])
play_store['Rating'] = pd.to_numeric(play_store['Rating'])
play_store['Reviews'] = pd.to_numeric(play_store['Reviews'])
play_store['Size'] = pd.to_numeric(play_store['Size'])
play_store['Price'] = pd.to_numeric(play_store['Price'])


play_store.info()
```

```python
# Number of Nan, Not NaN and Unique values in data frame

def Detailed_Data_Nan():
  Detailed_Data = pd.DataFrame(index = Df_review.columns)
  Detailed_Data['null values'] = Df_review.isna().sum()
  Detailed_Data['not null values'] = Df_review.notna().sum()
  Detailed_Data['unique values'] = Df_review.nunique()
  Detailed_Data['% of null values'] = (Df_review.isna().mean()*100).round(2)
  Detailed_Data['% of not null values'] = (Df_review.notna().mean()*100).round(2)

  return Detailed_Data

Detailed_Data_Nan()
```

```python
# visualizing Nan,Unique and Not_Nan values 

def Detailed_visual_Data_Nan():
  Detailed_visual_Data = pd.DataFrame(index = Df_review.columns)
  Detailed_visual_Data['null values'] = Df_review.isna().sum()
  Detailed_visual_Data['not null values'] = Df_review.notna().sum()
  Detailed_visual_Data['unique values'] = Df_review.nunique()

  percentage_of_Nan = pd.DataFrame(index = Df_review.columns)
  percentage_of_Nan['% of null values'] = (Df_review.isna().mean()*100).round(2)
  percentage_of_Nan['% of not null values'] = (Df_review.notna().mean()*100).round(2)

  plt.rcParams['figure.figsize'] = (10, 4)
  return Detailed_visual_Data.plot(kind='bar'), percentage_of_Nan.plot(kind='bar')

Detailed_visual_Data_Nan()

```
